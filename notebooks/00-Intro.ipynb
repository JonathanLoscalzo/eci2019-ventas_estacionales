### Descripción del problema
“Fiestas Dulces” es una empresa que entrega sus artículos en los puntos de venta a terceros (consignación). El producto insignia de la empresa se consume mayoritariamente en las fiestas a fin de año. El problema es que los artículos que comercializa son perecederos. La práctica usual es descartar lo que no se vende.

Para maximizar la ganancia “Fiestas Dulces” necesita minimizar dicho descarte. Una forma de hacer esto es conseguir buenas predicciones sobre la venta de sus productos en cada punto de venta.

El objetivo es predecir la cantidad de artículos que se venderá en cada punto de venta en diciembre de 2014.

### Descripción de los datos
Se proveen tres archivos: pos.csv, ventas.txt y envios.csv.
En pos.csv contiene información sobre los puntos de venta: tenemos información sobre porcentajes de edad, ingresos, densidad poblacional, porcentajes de  la población según estudios alcanzados, porcentaje de utilización de medios de transporte. 

En ventas.txt tenemos la fecha, el identificador del punto de venta y la cantidad de productos vendidos en dicho día. 

En envios.csv tenemos los envíos que se realizaron desde la central al punto de venta en cuestión, con esto podríamos saber la diferencia entre lo vendido y lo recibido.
